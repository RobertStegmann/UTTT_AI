```python
import matplotlib
import matplotlib.pyplot as plt
import scipy
import pandas as pd
import numpy as np
import math

data = pd.read_csv("./TestResults/heur_to_float.csv",names = ['Heurisitic Value', 'T/N'])
data.plot(kind='scatter',x='Heurisitic Value',y='T/N') # scatter plot

def fast_sig(x):
    return 0.5*((x/(150+abs(0.01*x))+1))

def sig(x):
    return 0.5*math.tanh(0.002*x)+0.5

heur = data['Heurisitic Value']
sig_heur = heur.apply( sig)
data.insert(2, "Sig_Heur", sig_heur, True)

data.plot(kind='scatter',x='Sig_Heur',y='T/N') # scatter plot
```
